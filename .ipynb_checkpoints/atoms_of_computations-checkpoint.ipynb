```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram
```

```python
n = 8
n_q = 8
n_b = 8
qc_output = QuantumCircuit(n_q,n_b)
```

```python
for i in range(n):
    qc_output.measure(i,i)
```

```python
qc_output.draw(output='mpl')
```

```python
result = execute(qc_output, Aer.get_backend('qasm_simulator')).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
qc_encode = QuantumCircuit(n)
```

```python
qc_encode.x(7)
```

```python
qc_encode.draw(output='mpl')
```

```python
qc = qc_encode + qc_output
```

```python
qc.draw(output='mpl')
```

```python
counts = execute(qc, Aer.get_backend('qasm_simulator')).result().get_counts()
```

```python
plot_histogram(counts)
```

```python
# Setting Age as Quantum Bits
qc_encode1 = QuantumCircuit(n)
# Binary of 21 - 00010101
qc_encode1.x(7)
qc_encode1.x(5)
qc_encode1.x(3)
qc_encode1.draw(output='mpl')
```

```python
# Plotting In Histogram
qc = qc_encode1 + qc_output
counts1 = execute(qc, Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts1)
```

```python
# Exploring CNOT
qc = QuantumCircuit(2,2)
qc.x(0)
qc.cx(0,1)
qc.measure(0,0)
qc.measure(1,1)
qc.draw(output='mpl')
```

```python
# Half Adder Circuit
qc = QuantumCircuit(4,2)
qc.x(0) # flipping it to 1
qc.x(1) # flipping it to 1
qc.barrier()
qc.cx(0,2)
qc.cx(1,2)
qc.ccx(0,1,3)
qc.barrier()
qc.measure(2,0)
qc.measure(3,1)
qc.draw(output='mpl')
```

```python
counts = execute(qc, Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```
