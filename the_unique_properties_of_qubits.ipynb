```python
# import qiskit and visualizations
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
# z measurement of a qubit
measure_z = QuantumCircuit(1,1)
measure_z.measure(0,0)
measure_z.draw(output='mpl')
```

```python
# x measurement of a qubit
measure_x = QuantumCircuit(1,1)
measure_x.h(0)
measure_x.measure(0,0)
measure_x.draw(output='mpl')
```

```python
qc_0 = QuantumCircuit(1)
qc_0.draw(output='mpl')
```

```python
# measurement on z
qc = qc_0 + measure_z
print("Results for z measurement")
counts = execute(qc, Aer.get_backend('qasm_simulator')).result().get_counts()
qc.draw(output='mpl')
```

```python
# measurement on z
plot_histogram(counts)
```

```python
# measurement on x
print("Results for x measurement")
qc = qc_0 + measure_x
counts = execute(qc, Aer.get_backend('qasm_simulator')).result().get_counts()
qc.draw(output='mpl')
```

```python
# measurement on x
plot_histogram(counts)
```

```python
# new circuit with hadamard 
qc_h = QuantumCircuit(1,1)
qc_h.h(0)
qc_h.draw(output='mpl')
```

```python
qc = qc_h + measure_z
qc.draw(output='mpl')
```

```python
# results with z measurement
counts = execute(qc, Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```
